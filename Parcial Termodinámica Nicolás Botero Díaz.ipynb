```python
v2 = 30 #Voluemn del gas en litros
temp_v2 = 27 #Temperatura del gas
aumento = 400  #aumento de la temperatura
```

```python
r0 = temp_v2 + 273 #300
r1 = round((aumento + temp_v2)+273,res)  #700
r2 = round((v2*r1/r0),res) #70

f0 = '\\frac{V_1}{T_1} = \\frac{V_2}{T_2}'
f1 = 'V_2 = \\frac{V_1T_2}{T_1} = \\frac{(%sL) (%s°K) }{%s °K}=%sL'%(v2,r1,r0,r2)

Markdown(f'''### Solución
Para los procesos isobáricos se puede emplear la ley de Charles que establece qué: 

$${f0}$$
Teniendo en cuanta lo anterior convertimos las temperaturas a °K para tener valores absolutos.

$$T_1 = {temp_v2} °C + 273°K = {r0} °K$$
$$T_2 = ({aumento} + {temp_v2})  °C + 273°K = {r1} °K$$

Ahora, despejando el $V_2$ y sabiendo que $V_1 = {v2} L$ se obtine:

$${f1}$$

Con ello, la variación es: 

$$\Delta V = V_2 - V_1 = {r2}L - {v2}L = {round(r2-v2,res)}L$$


''')
```

### Solución
Para los procesos isobáricos se puede emplear la ley de Charles que establece qué: 

$$\frac{V_1}{T_1} = \frac{V_2}{T_2}$$

Teniendo en cuanta lo anterior convertimos las temperaturas a °K para tener valores absolutos.

$$T_1 = 27 °C + 273 °K = 300 °K$$
$$T_2 = (400 + 27)  °C + 273 °K = 700 °K$$

Ahora, despejando el $V_2$ y sabiendo que $V_1 = 30 L$ se obtine:

$$V_2 = \frac{V_1T_2}{T_1} = \frac{(30L) (700°K) }{300 °K} = 70 L$$

Con ello, la variación es: 

$$\Delta V = V_2 - V_1 = 70L - 30L = 40L$$

$$P_1V_1^\alpha =P_2V_2^\alpha$$

$$\alpha = \frac{i+2}{i}$$

donde i es el tipo de gas que se trabaja y como el neon es monoatómico i es 3.

$$\alpha = \frac{3 + 2}{3} = \frac{5}{3}$$

$$(4Pas)(64L)^{\frac{5}{3}}=(128Pas)(V_2)^{\frac{5}{3}}$$

$$V_2^{\frac{5}{3}}=\frac{(2^2Pas)(2^6L)^{\frac{5}{3}}}{2^7Pas}$$
$$V_2^{\frac{5}{3}}=\frac{(2^2Pas)(2^10L)}{2^7Pas}$$

$$V_2 = \sqrt[5]{(2^5 L)}^3 = 8L$$

## Evaluación Cuantitativa de Riesgos

En la Evaluación Cuantitativa de Riesgos, se utilizan índices y cálculos específicos para determinar la magnitud del riesgo en una ubicación determinada. Esto implica evaluar el Riesgo Individual Específico de Ubicación (LSIR), que representa la probabilidad anual de muerte o lesiones graves para una persona en las cercanías de una amenaza. 

#### Definiciones

- **Índices de Riesgo (RI):** Son números o tabulaciones de números que se correlacionan con la magnitud del riesgo. (CCPS, 1999).

- **Riesgo Individual Específico de Ubicación (LSIR):** LSIR es el riesgo anual de muerte o lesiones graves para una persona en la cercanía de una amenaza, suponiendo una exposición continua durante todo el año. Se calcula utilizando la siguiente fórmula:

$$LSIR_{x, y} = \sum_{i=1}^{n} LSIR_{x, y, i}$$

Donde:
  - $LSIR_{x, y}$ representa las posibilidades de fatalidad por año en la ubicación geográfica $x, y$.
  - $LSIR_{x, y, i}$ representa las posibilidades de fatalidad por año en la ubicación geográfica $x, y$ debido al caso de resultado $i$ (posibilidades de fatalidad por año o año 1).
  - $n$ es el número total de casos de resultado de incidentes considerados en el análisis.

Y:

$$LSIR_{x, y, i} = f_i P_{x, y, i}$$

Donde:
  - $f_i$ es la frecuencia del caso de resultado de incidente $i$ (año 1).
  - $P_{x, y, i}$ es la probabilidad de que el caso de resultado de incidente $i$ resulte en una fatalidad en la ubicación $x, y$.

### Cálculo de Probabilidad de Muerte

La probabilidad de muerte $P_{x, y, i}$ debido a la exposición a una concentración específica de una sustancia en una ubicación determinada se calcula utilizando la función probit, como se describe en el Capítulo 6:

$$P_{x, y, i} = a + b \ln(C_{n_{x, y, i}} t_{x, y, i})$$

Donde $a$, $b$, $C_{n_{x, y, i}}$ y $t_{x, y, i}$ son parámetros específicos de la sustancia y la ubicación.

Este enfoque difiere del enfoque generalizado por el CCPS.

En una ubicación específica, el LSIR total $LSIR_{x, y}$ se compone de tres partes: LSIR debido al impacto tóxico (TOX), LSIR debido al impacto térmico (TH), y LSIR debido al impacto de sobrepresión (OP):

$$LSIR_{x, y} = LSIR_{x, y, TOX} + LSIR_{x, y, TH} + LSIR_{x, y, OP}$$

Estos valores LSIR se pueden tabular y representar en iso-frecuencias, lo que permite visualizar el riesgo total y el riesgo asociado a cada tipo de peligro.

#### Individual Risk Per Annum (IRPA)

El IRPA (Riesgo Individual por Año) se refiere al riesgo anual de muerte o lesiones graves para una persona cercana a una amenaza, teniendo en cuenta la fracción del año real que la persona pasa en el lugar. Se calcula de la siguiente manera:

$$IRPA = f \cdot LSIR$$

Donde $f$ es la fracción anual que una persona pasa en el lugar. Se representan también contornos de frecuencia para los valores de IRPA, basados en la distribución de personal.

#### Maximum Individual Risk

El Riesgo Individual Máximo se refiere al riesgo para la persona o personas expuestas al riesgo más alto en una población expuesta.

#### Average Individual Risk

El Riesgo Individual Promedio es el promedio de todas las estimaciones de riesgo individual en una población definida y se calcula de la siguiente manera:

$$LSIR_{AV} = \frac{\sum_{x, y} LSIR_{x, y} \cdot \Pi_{x, y}}{\sum_{x, y} \Pi_{x, y}}$$

Donde $\Pi_{x, y}$ representa el número de personas en la ubicación $x, y$.

#### Potential Loss of Life (PLL) or Rate of Death (ROD)

El Potencial de Pérdida de Vida (PLL) o Tasa de Muerte (ROD) se refiere al número esperado de fatalidades dentro de una población específica por año, tanto en tierra como en mar. Se calcula de la siguiente manera:

- Para la costa (On-shore):

$$PLL = IRPA \cdot POS = LSIR \cdot POS \cdot f_{OS}$$

- Para fuera de la costa (Off-shore):

$$PLL = IRPA \cdot POB = LSIR \cdot POB \cdot f_{OB}$$

Donde:
- $POS$ es el número de personas en el lugar.
- $POB$ es el número de población a bordo.
- $f_{OB}$ y $f_{OS}$ son las respectivas fracciones del tiempo invertido.

#### Fatal Accident Rate (FAR)

La Tasa de Accidentes Mortales (FAR) es una medida de riesgo individual expresada como el número estimado de fatalidades por cada 100 millones de horas de exposición (aproximadamente 1000 vidas laborales de empleados). Se calcula de la siguiente manera:

$$FAR = \frac{LSIR_{AV}}{11,415}$$

Donde 11,415 es la relación de 100 millones de horas a las horas en un año (8760).

Estos conceptos se utilizan para evaluar y comprender el riesgo en entornos industriales y se aplican a diversas actividades para garantizar la seguridad de las personas.

#### 12.3 PLL a partir de la Curva FN

El diagrama FN (Frecuencia vs. Número de Fatalidades) define matemáticamente dos Potenciales de Pérdida de Vida (PLL), a saber:
- El PLL correspondiente a la línea superior.
- El PLL correspondiente a la línea inferior.

Según la definición de frecuencia acumulativa, para la línea superior:

$$f(N) = F(N)-F(N + 1) \quad (12.21)$$

Para la línea roja (superior):

$$F(N) = \frac{1}{10\times N} \quad (12.22)$$

Por definición de PLL:


$$PLL_{superior} = \int_{1}^{N_{\text{max}}} f(N) \, dN = 0.1 \times \int_{1}^{N_{\text{max}}} \left[ \frac{1}{N}-\frac{1}{N+1} \right] dN= 0.1 \times \left[\ln{(N)} - \ln{(N+1)} \right]^{N_{max}}_{1}$$

**Solución:**

$$PLL_{superior} = \int_{1}^{70} f(N) \, dN = 0.1 \times \int_{1}^{70} \left[ \frac{1}{N}-\frac{1}{N+1} \right] dN= 0.1 \times \left[\ln{(N)} - \ln{(N+1)} \right]^{70}_{1}$$

$$=0.1\times \left[ \ln{(70)} - \ln{(71)} \right] - 0.1 \times \left[ \ln{(1)} - \ln{(2)} \right] = 0.0679y^{-1}$$

#### 12.4 ALARP Demonstration—ICAF Method

La banda central del modelo ALARP identifica una zona condicional donde el riesgo individual y la región FN son aceptables, siempre y cuando las medidas de mitigación de riesgos adoptadas generen una desproporción entre los costos adicionales y los beneficios esperados correspondientes en términos de mejora de la seguridad. Una técnica validada para determinar si una reducción de riesgo es razonablemente practicable es el ICAF (Costo Implícito de Evitar una Fatalidad).

Una fórmula posible para el ICAF es la siguiente:


$$ICAF = \frac{\text{COST}}{\text{PLT} \cdot \Delta PLL} \quad (12.25)$$

Donde:
- COST es el costo de implementar la medida.
- PLT es la vida útil de la planta.
- ΔPLL es la reducción del PLL.

Se muestra aquí un ejemplo de la aplicación del método ICAF, junto con algunos parámetros típicos adoptados con el método.


#### Ejemplo 5 

**Cálculo del ICAF para la ZONA ALARP para Trabajadores**

Se debe reducir el Potencial de Pérdida de Vida (PLL) de \(1.5 \times 10^3\) por año a un tercio. La vida útil de la planta es de 30 años. Encuentra el ICAF (Costo Implícito de Evitar una Fatalidad) en la ZONA ALARP para los trabajadores si el valor de una fatalidad estadística es de 1,000,000 dólares estadounidenses (HSE, 2002).

**Solución:**

Según el HSE (2002), las líneas de borde inferior y superior de los factores de desproporción de las zonas ICAF y ALARP son respectivamente 1 y 10 (Figura 12.7).

El COST (costo) correspondiente al factor de desproporción = 1 es:

$$ICAF_L \times PLT \times \Delta PLL = 1,000,000 \times 30 \times (1.5-0.5)\times 10^{-3} = 30,000 \text{ dólares} \quad (12.26)$$

El COST correspondiente al factor de desproporción \(= 10\) es:

$$ICAF_U \times PLT \times \Delta PLL = 10,000,000 \times 30 \times (1.5-0.5)\times 10^{-3} = 300,000 \text{ dólares} \quad (12.26)$$

Este es un ejemplo. Es importante señalar que las reglas financieras deben ser establecidas por la política corporativa. En estos casos, se pueden utilizar métodos semicuantitativos, como las matrices ALARP, para establecer el nivel de cumplimiento y verificar si es necesario adoptar un enfoque más seguro.
